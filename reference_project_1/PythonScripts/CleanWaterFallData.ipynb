```python
import pandas as pd
data = pd.read_csv("../Data/employer_salary_emp_count&p2.csv")
```

```python
# df = df.sort_values(['employerId', 'week'],ignore_index=True)
```

```python
df = data.sort_values(['employerId', 'week', 'participant_count'], ascending=[True, True, False])
df = df.groupby(['employerId', 'week']).apply(lambda x: x.loc[x['participant_count'].idxmax()]).reset_index(drop=True)
df
```

```python
df.size
len(df)
```

```python
df["prevCount"] = 0
i = 0
while i < len(df):
    for j in range(int(df["week"].max())):
        if(i >= len(df)):
            break
        if(df.loc[i,"week"]!=1): # skip first for each empolyer 
#             print(df.loc[i-1]["participant_count"])
            df.at[i, "prevCount"] = df.at[i-1, "participant_count"]        
        i+= 1;
    
df
    
```
